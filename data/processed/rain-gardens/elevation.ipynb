```python
import geopandas as gpd
import pandas as pd
```

```python
slope = pd.read_csv(
    "/home/zagreus/leap-hackathon-flushing/data/original/elevation/urban_Mean_slope.csv"
)
tracts = gpd.read_file(
    "/home/zagreus/leap-hackathon-flushing/data/nyc_census_tracts/nyc_census_tracts.shp"
)
print(tracts.columns)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, figsize=(8, 8))

tracts.plot(
    column="Slope",
    cmap="viridis",
    linewidth=0.2,
    edgecolor="gray",
    legend=True,
    ax=ax,
)

ax.set_title("NYC Census Tracts  Mean Slope")
ax.set_axis_off()

plt.show()
```
